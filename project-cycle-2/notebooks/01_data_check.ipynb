# 01 Data Check

## Research Question
1. Is the proportion of students who currently smoke different from 0.20?
2. Is the mean height of students different from 1.70 meters?

## Variables
- Proportion variable: `CurrentCigaretteUse`
- Continuous variable: `HowTallAreYouWithoutShoesInMeters`


## Variable Definition and Coding

### Proportion variable
- Variable name: `CurrentCigaretteUse`
- Meaning: current cigarette use
- Recoding rule:
  - success = codes 2–7
  - failure = code 1
- Missing or invalid values are removed before analysis.

### Continuous variable
- Variable name: `HowTallAreYouWithoutShoesInMeters`
- Meaning: height without shoes in meters
- Missing values are removed before analysis.


```python
import pandas as pd

# Load raw data
df = pd.read_csv('../data/raw/YRBS_2007.csv')

print('=== Dataset shape ===')
print(df.shape)

print('\n=== First 5 rows ===')
display(df.head())

print('\n=== Data info ===')
df.info()

print('\n=== Missing values ===')
display(df.isnull().sum().sort_values(ascending=False).head(20))
```

```python
# Data check for CurrentCigaretteUse
smoke_raw = df['CurrentCigaretteUse']

print('=== CurrentCigaretteUse original frequencies (including NaN) ===')
display(smoke_raw.value_counts(dropna=False).sort_index())

def recode_smoking(x):
    if pd.isna(x):
        return pd.NA
    elif x in [2, 3, 4, 5, 6, 7]:
        return 1
    elif x == 1:
        return 0
    else:
        return pd.NA

df['CurrentCigaretteUse_Bin'] = df['CurrentCigaretteUse'].apply(recode_smoking)
smoke_bin = df['CurrentCigaretteUse_Bin'].dropna()

print('\n=== Recoded smoking frequencies ===')
display(smoke_bin.value_counts(dropna=False).sort_index())

print('\n=== Recoded smoking proportions ===')
display(smoke_bin.value_counts(normalize=True).sort_index())

print(f'Final sample size for smoking analysis: {len(smoke_bin)}')
```

### Short observation
- The original smoking variable should be checked before recoding.
- The final binary variable clearly defines success and failure for proportion inference.


```python
# Data check for height
height = df['HowTallAreYouWithoutShoesInMeters'].dropna()

print('=== Height sample size ===')
print(len(height))

print('\n=== Height summary ===')
display(height.describe())

print('\n=== Possible invalid values check ===')
display(height[(height <= 0) | (height > 3)].head())

print(f'Final sample size for height analysis: {len(height)}')
```

### Short observation
- The height variable is continuous, so it is appropriate for mean inference.
- Missing values are removed before analysis.
- Any impossible values should be checked and documented if found.
